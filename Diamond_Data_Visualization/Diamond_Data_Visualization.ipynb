The dataset based on different types diamonds and their respective charactertics 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('.\diamonds.csv')
```

A set of plots between all the features on the dataset.

```python
sns.pairplot(data=df)
```

#### The rest of the graphs are self-explanatory. The keyword after "sns" defines the type of the graph and labels x and y correspond of the features of the dataset (df) used in the graphs as x and y axis respectively.

```python
x = df['price']
sns.distplot(x)
```

```python
x = df['depth']
sns.distplot(x)
```

```python
x = df['carat']
sns.distplot(x)
```

```python
sns.kdeplot(x)
sns.kdeplot(x, bw=.2, label="bw: 0.2")
sns.kdeplot(x, bw=2, label="bw: 2")
plt.legend();
```

```python
sns.scatterplot(x='price',y='carat',data=df,hue='price')
```

```python
sns.jointplot('price','carat',df,kind='hex')
```

```python
sns.scatterplot(x='price',y='clarity',data=df,hue='price')
```

```python
sns.scatterplot(x='price',y='depth',data=df,hue='price')
```

```python
sns.jointplot('price','depth',df,kind='kde')
```

```python
sns.countplot('cut',data=df)
```

```python
sns.countplot('color',data=df)
```

```python
sns.countplot('clarity',data=df)
```
